```
import openpyxl
from string import punctuation
from nltk.corpus import stopwords
```

```
def copytoExcel(filename , row , column , value):

        wrkbk = openpyxl.load_workbook(filename)
        exceldata = {}
        sh = wrkbk.active
        c1 = sh.cell(row=row, column=column)

        # writing values to cells
        c1.value =  value

       # c2 = sheet.cell(row=1, column=2)
       # c2.value = "RAI"
        wrkbk.save( filename)
        wrkbk.close()

def ReadfromExcel(filename):

    # load excel with its path
        wrkbk = openpyxl.load_workbook(filename)
        exceldata = {}
        sh = wrkbk.active
        appendtext = ''
        # iterate through excel and display data
        for i in range(1, sh.max_row+1):
          #  print("\n")
           # print("Row ", i, " data :")
            appendtext = ''
            for j in range(1, sh.max_column + 1):
                cell_obj = sh.cell(row=i, column=j)
              #  exceldata[i].append(cell_obj.value)
                #if(cell_obj.value )
                appendtext = appendtext + str(cell_obj.value) + ";"

                #print(appendtext)
            dict1(exceldata, i, appendtext)
        wrkbk.close();

        return exceldata


def clean_doc(doc):
        # split into tokens by white space
        tokens = doc.split()
        # remove punctuation from each token
        table = str.maketrans('', '', punctuation)
        tokens = [w.translate(table) for w in tokens]
        # remove remaining tokens that are not alphabetic
        tokens = [word for word in tokens if word.isalpha()]
        # filter out stop words
        stop_words = set(stopwords.words('english'))
        tokens = [w for w in tokens if not w in stop_words]
        # filter out short tokens
        tokens = [word for word in tokens if len(word) > 1]
        return tokens

def removeBrackets(str):
    str = str.replace('(','')
    str = str.replace(')','')
    # str = str.replace('=',',')
    print(str)
    return str

def splitString(str):
	splittedString = str.split('=')
	for i in range(len(splittedString)):
		splittedString[i] = splittedString[i].strip()
	# print(splittedString)
	return splittedString

def Ques1Ans(str1, str2):
    str1 = str1.replace("QAns1" , str2)
    return str1

# exceldata1 = ReadfromExcel(r"final3.xlsx")
ContextAndQuestionmap = {}
```
